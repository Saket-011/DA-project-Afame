```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing HR Data as csv
Data_1=pd.read_csv("/content/HR Data.csv")
Data_1
```

```python
#Check the columns to see their relevance
Data_1.columns
```

```python
#Let us drop those columns whoare irrelevent
Data_2=Data_1.drop(["EmployeeCount","EmployeeNumber","Over18","StandardHours"],axis=1)
Data_2
```

```python
#Change the names of coloumns
Data_3=Data_2.rename(columns={'BusinessTravel':'Travel_Frequency','TotalWorkingYears':'Years_of_Experience'})
Data_3.columns
```

```python
#Lets check the count of NaN values in each row
Data_3.isnull().sum()
```

Since no NaN values are present , our data is clean.

```python
#It can be seen that some columns have yes or no instead of 1/0.
mapping = {'Yes': 1, 'No': 0}

Data_3['Attrition'] = Data_3['Attrition'].map(mapping)

Data_3
```

```python
#Some coloumns have String Data instead of numerical ones
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Fit and transform the data
Data_3['Travel_Frequency'] = label_encoder.fit_transform(Data_3['Travel_Frequency']) # 2->Travel Rarely ; 1->Travel Frequently.

Data_3
```

```python
#Similar approaches can be used to other column containing String Data

Data_3['Department'] = label_encoder.fit_transform(Data_3['Department']) #2-> Sales; 1-> R&D

Data_3['EducationField'] = label_encoder.fit_transform(Data_3['EducationField']) #1->Life Sciences ; 3->Medical; 4->Others

Data_3['Gender'] = label_encoder.fit_transform(Data_3['Gender']) #1->Male ; 0->Male

Data_3
```
